```python
# Goal: Use Python to implement linear regression with gradient descsent and fit a line to synthetic data
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create synthetic data: y = 3x + 4 + noise
np.random.seed(42) # for reproducibility
X = 2 * np.random.rand(100, 1) # 100 values between 0 and 2
y = 3 * X + 4 + np.random.rand(100, 1)
```

```python
# plot the data
plt.scatter(X,y)
plt.title("Synthetic Linear Data")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
# initialize weights (slope) and bias (intercept)
m = 0
b = 0
```

```python
# set hyperparameters
learning_rate = 0.05
epochs = 1000 # no. of iterations
n = len(X) # no. of trainig values
```

```python
# track loss over time
loss_history = []
```

```python
# gradient descsent loop
for i in range(epochs):
    # predict current values
    y_pred = m * X + b

    # compute gradients
    error = y_pred - y
    dm = (2/n) * np.sum(X * error)
    db = (2/n) * np.sum(error)

    # update parameters
    m -= learning_rate * dm
    b -= learning_rate * db

    # calculate and store loss (MSE)
    loss = (1/n) * np.sum(error**2)
    loss_history.append(loss)

    # print updates every 100 iterations
    if i % 100 == 0:
        print(f"Epoch {i}: Loss = {loss:.4f}, m = {m:.4f}, b = {b:.4f}")
```

```python
# visualize the results
```

```python
# final fitted line
plt.scatter(X, y, label='Data')
plt.plot(X, m * X + b, color='red', label='Fitted Line')
plt.title("Linear Regression via Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

```python
# plot loss over epochs
plt.plot(range(epochs), loss_history)
plt.title("Loss over Epochs")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```
